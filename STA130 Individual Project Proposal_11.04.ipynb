### Research Question #1
How does the frequency of online socialization relate to perceived loneliness among university students? With the growing prevalence of online interactions, especially among younger adults, this study seeks to explore whether students who frequently socialize online experience more or less emotional, social, and situational loneliness than those who engage less in digital interactions. </p>


### Variables
1. Online Socialization Frequency:</br>
- Measures how often students engage in online social activities, such as messaging, social media use, and virtual hangouts.</br>
- Online socialization frequency is the independent variable, representing the extent of each student's digital interactions.</p>

Visualizations:
- A histogram to visualize the distribution of online socialization frequency among students.</br>
- A scatter plot showing the correlation between online socialization and perceived loneliness.</p>

2. Perceived Loneliness:</br>
    1. Emotional Loneliness:
    - Items capturing feelings of emotional isolation, such as “I experience a general sense of emptiness” and “I often feel rejected.”
    - Emotional loneliness reflects deep-seated feelings of isolation and unmet emotional needs, which may or may not be alleviated by online social interactions.
    2. Social Loneliness: 
    - Items capturing the sense of connectedness to others, like “There are enough people I feel close to” and “I miss having people around.”
    - Social loneliness measures one’s sense of closeness or belonging and is likely influenced by online social interactions.
    3. Situational Loneliness:
    - Items describing specific social needs, such as missing meaningful conversation, physical touch, or companionship.
    - Situational loneliness provides context on the types of social interactions students feel are lacking when lonely.</p>
    
Visualizations:</br>
- A box plot to compare loneliness scores across various levels of online socialization.
- A bar chart displaying average loneliness scores for students grouped by different online interaction frequency categories.
- Heatmaps showing correlations among loneliness subcategories and online socialization.</p>

3. Control Variables:
- Demographics and lifestyle factors such as age, year of study, and **in-person socialization frequency.**
- Controlling for these variables helps isolate the effect of online socialization on loneliness by accounting for other factors that may affect students' feelings of loneliness.</p>

Visualizations: </br>
- Summary tables and stratified box plots showing how loneliness varies across these control variables within each online socialization category.</p>

### Analysis
This study will calculate composite scores for each loneliness subcategory by averaging responses to relevant items (e.g., “Emotional Loneliness” based on feelings of emptiness and rejection). Each loneliness subcategory will be analyzed separately in relation to online socialization frequency using regression models, with control variables added to account for confounding factors. Before running the regressions, assumptions of linearity and independence will be checked to confirm model suitability.</p>

### Hypothesis and Expected Results
**Hypothesis:** Higher online socialization frequency is associated with lower social loneliness but may have limited effects on emotional and situational loneliness. This hypothesis stems from the idea that online interactions may help students feel connected socially but might not fulfill deeper emotional needs or specific social experiences.</p>

**Expected Results:** If online socialization is found to primarily reduce social loneliness but not emotional loneliness, it may suggest that digital interactions support a sense of belonging but not necessarily emotional fulfillment. This insight could inform recommendations encouraging balanced in-person interactions to help meet students’ emotional and situational social needs.</p>

By distinguishing between different types of loneliness, this analysis provides a nuanced view of how online socialization impacts various aspects of loneliness among students. Such findings could guide mental health initiatives that emphasize in-person interactions for more comprehensive social well-being.</p>

### Research Question #2
How does engaging in conversations with strangers impact the well-being and loneliness of university students? In modern society, interactions with strangers (such as casual conversations with peers, neighbors, or even random encounters) are common yet often overlooked as sources of social connection. This study aims to examine whether such interactions contribute positively or negatively to students' overall sense of well-being and feelings of loneliness.</p>

### Variables
1. Frequency of Talking to Strangers: (CONNECTION_activities_greeted_stranger_p3m)
- This variable measures how often students engage in conversations with strangers, such as classmates, neighbors, or people they encounter in public settings.
- The frequency of talking to strangers will serve as the independent variable and represents the extent to which students engage in spontaneous, casual social interactions.</br>
Suggested Visualizations:
- A histogram to show the distribution of frequency of talking to strangers among students.
- A scatter plot to assess the relationship between talking to strangers and well-being/loneliness measures.</p>

2. Well-Being Indicators:
- Variables measuring different aspects of well-being, such as self-rated happiness, satisfaction with social life, and stress levels.
- These indicators will act as dependent variables, allowing us to observe if talking to strangers has a beneficial or detrimental effect on students’ well-being.</br>
Suggested Visualizations:
- Box plots comparing well-being scores across different frequencies of stranger interactions.
- Line charts tracking well-being indicators against increasing frequencies of talking to strangers.</p>

3. Loneliness Indicators:
- Loneliness will be measured using scales like the UCLA Loneliness Scale, including items that assess social exclusion, companionship, and isolation (e.g., "How often do you feel left out?").
-  These indicators serve as the primary dependent variables, capturing different dimensions of students' perceived loneliness. </br>
Suggested Visualizations:
- Box plots comparing loneliness scores across different frequencies of talking to strangers. 
- Heatmaps displaying correlations between talking to strangers and various loneliness indicators. </p>

4. Control Variables:
- Demographic and situational factors, such as age, personality type (e.g., introvert vs. extrovert), and regularity of in-person social interactions.
- Controlling for these variables will help clarify whether any effects observed are specifically due to interactions with strangers or could be influenced by other factors.</br>
Suggested Visualizations: 
- Summary tables and stratified scatter plots to display loneliness or well-being scores across control variables.</p>

### Analysis
To assess the impact of talking to strangers on well-being and loneliness, regression models will be used with well-being and loneliness as dependent variables and frequency of talking to strangers as the primary predictor. Control variables will be added to account for personal and situational factors. Prior to running the analysis, assumptions like linearity and independence will be checked to confirm model suitability. Additionally, interactions between personality type and frequency of stranger interactions may be explored to see if effects vary based on personal disposition (e.g., introversion vs. extroversion).</p>

### Hypothesis and Expected Results
**Hypothesis:** Engaging in conversations with strangers is expected to have a positive association with well-being (e.g., higher happiness and satisfaction scores) and a negative association with loneliness (lower loneliness scores). This hypothesis is based on the idea that casual interactions can create a sense of connection and reduce feelings of isolation, though the effects might vary by personality type.</p>

**Expected Results:** If a positive relationship between talking to strangers and well-being is found, it could suggest that casual interactions provide meaningful social engagement, even if they’re brief or superficial. Alternatively, if certain personality types (such as introverts) experience increased stress or loneliness from such interactions, this may highlight the need for personalized approaches in encouraging social interaction for well-being.</p>

This study can provide valuable insights into the potential benefits of casual social interactions, offering guidance on whether initiatives that encourage such interactions could enhance student well-being. These findings could also contribute to campus programs or mental health strategies that promote healthy social habits among students.</p>

### Research Question #3
Does living a balanced lifestyle that includes diverse social activities decrease loneliness among university students? This study examines if a variety of social interactions, such as engaging with friends, family, community, and participating in both in-person and virtual activities, can help mitigate feelings of loneliness.</p>

### Variables:
1. Lifestyle Balance Score:(independent variable)
- This composite score represents the extent to which students engage in a variety of social activities, measured -through their participation in activities such as:
    1. Social Activities with Friends and Family: e.g., “visiting with friends,” “having a coffee with someone,” “going for a walk with someone.”
    2. Virtual Interactions: e.g., “extended conversations via text,” “video chat with friends or family,” “online gaming.”
    3. Community Engagement: e.g., “volunteering in the community,” “helping neighbors with chores,” “participating in group exercise.”
- A high balance score would indicate students are participating in a range of activities, suggesting a more well-rounded social life.. </br>
Suggested Visualizations: </br>
A histogram displaying the distribution of balanced lifestyle scores.
A bar chart showing average loneliness scores across different levels of lifestyle balance.</p>

2. Loneliness Indicators: (dependent variable)
- This includes specific measures of loneliness, such as:
    1. Emotional and Social Loneliness: Measured through items like “I experience a general sense of emptiness” and “There are enough people I feel close to.”
    2. UCLA Loneliness Scale: Including items such as “How often do you feel left out?” and “How often do you feel isolated from others?”
- These indicators provide insights into different dimensions of loneliness, allowing us to assess how lifestyle balance impacts both emotional and social aspects of loneliness. </br>
Suggested Visualizations: </br>
- Box plots comparing loneliness levels across groups with different lifestyle balance scores.
- Heatmaps showing correlations between various activities (e.g., virtual interactions vs. in-person activities) and loneliness indicators .</p>

3. Types of Social Activities (Social Interaction Types:): </br>
    1. Casual Interactions: e.g., “greeting a stranger,” “checking in with friends via text.”
    2. In-Person Social Activities: e.g., “playing a board game with others,” “meeting someone for coffee,” “attending religious gatherings.”
    3. Virtual Connections: e.g., “video chats with friends or family,” “participating in online discussion groups,” “playing online games with others.”
- Separating these activities allows for analysis of whether specific types (in-person vs. virtual) are more effective at reducing loneliness. </br>

Suggested Visualizations:
- Scatter plots showing relationships between each activity type (in-person, virtual, etc.) and loneliness.
- A line graph tracking loneliness indicators against varying levels of engagement in specific activity types.</p>

4. Control Variables:
- Variables such as age, year of study, and preferences for certain types of socialization (e.g., preference for socializing with family vs. friends).
- Including these controls ensures observed relationships are due to lifestyle balance itself, rather than individual preferences or demographic factors.</br>
Suggested Visualizations: Stratified box plots displaying loneliness scores across control variables within different levels of lifestyle balance.</p>

### Analysis
To examine the association between a balanced lifestyle and loneliness, a multiple regression analysis will be conducted with loneliness indicators as the dependent variable and the balanced lifestyle score as the main predictor. Individual social activities will also be analyzed separately to see which activities are most associated with decreased loneliness. Control variables will be included to adjust for demographic factors, and interaction terms may be explored to understand if certain personality types or lifestyle preferences influence the relationship between balance and loneliness.

### Hypothesis and Expected Results
**Hypothesis:** A more balanced lifestyle, with frequent engagement across diverse social activities, is associated with lower loneliness. Specifically, in-person and community-oriented activities are expected to have a stronger association with reduced loneliness than virtual interactions alone.</br>

**Expected Results:** If a balanced lifestyle score is inversely related to loneliness indicators, this would suggest that a diverse social life, incorporating a variety of activities, provides emotional and social support that reduces loneliness. Results highlighting the greater impact of in-person and community activities could emphasize the importance of face-to-face interactions over virtual-only engagements.</br>

This study would provide valuable insights into how different aspects of a balanced lifestyle affect loneliness, offering guidance for students and universities on promoting holistic, diverse social activities to improve mental health and social well-being.</br>
