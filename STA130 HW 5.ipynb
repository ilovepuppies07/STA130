# Statistical Analysis of Fishers Tea Experiment: STA130 Student Sample

## Problem Introduction
The original Fishers Tea Experiment was designed to test whether Dr. Muriel Bristol could distinguish between cups of tea where milk was poured first and those where tea was poured first. In our version of this experiment, we selected a random sample of 80 STA130 students, each tasting one cup of tea and guessing which was poured first. Out of these 80 students, 49 correctly identified whether milk or tea was poured first. The purpose of this analysis is to determine whether the students success rate is better than random guessing, or if it can be explained by chance alone.

## Relationship to Fisher's Original Experiment
In Fisher's original experiment, Dr. Bristol claimed she could tell the difference, and the sample size was very small: only 8 cups of tea were tested. The experiment was highly personalized, focusing on one individuals claim. In contrast, the current experiment expands the scope by involving 80 students from the STA130 course, making the parameter of interest less personalized and more abstract, as we aim to assess the general ability of a group to distinguish the tea-pouring order.

## Null Hypothesis (H) and Alternative Hypothesis (H)

- **Null Hypothesis (H):** The students are guessing randomly, and the probability of correctly identifying the pouring order is 0.5.
  
  *Formal version:* $H_0: p = 0.5$, where $p$ is the proportion of students who can correctly identify the pouring order.
  
  *Informal version:* The students are just guessing, so their success rate will be close to 50%, as expected by chance. </p>

- **Alternative Hypothesis (H):** The students have some ability to correctly identify the pouring order better than chance.

  *Formal version:* $H_1: p > 0.5$.
  
  *Informal version:* The students are able to distinguish between milk-first and tea-first pouring better than random guessing.

## Quantitative Analysis

- **Test Statistic:** The observed statistic is the proportion of correct identifications, which is:

  $\hat{p} = \frac{49}{80} = 0.6125$

This represents the sample proportion of students who correctly identified the pouring order. We will test this against the null hypothesis that the students are guessing, implying $p = 0.5$.

## Simulation Approach
To test whether the result of 49 correct answers could occur by chance, we simulated a sampling distribution under the null hypothesis $H_0$, assuming students are randomly guessing.

## Methodology:
- We simulated 10,000 repetitions of 80 students each, randomly guessing the pouring order. The proportion of correct answers in each trial was recorded.
- We then compared the observed proportion (0.6125) to the distribution of simulated proportions to calculate the probability (p-value) of observing a result at least as extreme as ours under the null hypothesis.

```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_students = 80
p_null = 0.5  # Null hypothesis probability
n_simulations = 10000

# Simulate 10,000 trials of 80 students guessing randomly
simulated_data = np.random.binomial(n_students, p_null, n_simulations) / n_students

# Calculate the observed proportion of correct answers
observed_proportion = 49 / 80

# Calculate the p-value (proportion of simulated results greater than or equal to observed)
p_value = np.mean(simulated_data >= observed_proportion)
p_value
```


## Results:
After running the simulation, the calculated p-value was approximately **0.037**. This means that, assuming students were randomly guessing, there is only about a 3.7% chance of observing 49 or more correct guesses out of 80.

## Findings and Discussion
The p-value of 0.037 is less than the common significance level of 0.05, which suggests that we have evidence to reject the null hypothesis. This implies that it is unlikely the result of 49 correct answers out of 80 is due to random guessing alone. Therefore, we conclude that the STA130 students might indeed have some ability to distinguish the pouring order of tea.

### Supporting Visualization
A histogram of the simulated proportions under the null hypothesis would show that the observed proportion of 0.6125 lies in the tail of the distribution, indicating that it is an unlikely result if students were just guessing.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
n_students = 80
p_null = 0.5  # Null hypothesis probability
n_simulations = 10000

# Simulate 10,000 trials of 80 students guessing randomly
simulated_data = np.random.binomial(n_students, p_null, n_simulations) / n_students

# Observed proportion of correct answers
observed_proportion = 49 / 80

# Plot histogram of simulated data
plt.hist(simulated_data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f'Observed Proportion: {observed_proportion:.4f}')
plt.title('Simulated Proportions Under H (Random Guessing)')
plt.xlabel('Proportion of Correct Guesses')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Conclusion Regarding the Null Hypothesis
Based on the quantitative analysis, we reject the null hypothesis that the students are randomly guessing. The observed proportion of 0.6125, with a p-value of 0.037, suggests that the STA130 students show some ability to distinguish whether milk or tea was poured first, better than what would be expected by random chance.
